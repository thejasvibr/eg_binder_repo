```python
from ipywidgets import *
import matplotlib.pyplot as plt
import numpy as np 
```

```python
%matplotlib notebook
```

```python

x = np.linspace(0, 2 * np.pi)
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
line, = ax.plot(x, np.sin(x))

def update(w = 1.0, g=np.arange(0,3.0)):
    line.set_ydata(np.sin(w * x)*g)
    fig.canvas.draw_idle()

interact(update);
```
